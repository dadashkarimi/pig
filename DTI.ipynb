```python
import nibabel as nib
import matplotlib.pyplot as plt
from dipy.data import fetch_bundle_fa_hcp
import urllib.request
import os

from utils import *
from help import *

# Fetch the data (this will give us the URL and checksum)
files, folder = fetch_bundle_fa_hcp()

# Extract the URL for the FA data
fa_url = files['hcp_bundle_fa.nii.gz'][0]
fa_filename = 'hcp_bundle_fa.nii.gz'

# Download the file if it's not already present
if not os.path.exists(fa_filename):
    print(f"Downloading {fa_filename}...")
    urllib.request.urlretrieve(fa_url, fa_filename)

# Load FA image
fa_img = sf.load_volume(fa_filename)
# fa_data = fa_img.get_fdata()

ne.plot.volume3D(fa_img,cmaps=['tab20c'])

```
